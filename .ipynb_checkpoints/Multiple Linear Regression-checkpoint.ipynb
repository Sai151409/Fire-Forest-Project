# Multiple Linear Regression

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
califorina = fetch_california_housing()
```

```python
X = pd.DataFrame(data = califorina.data, columns=califorina['feature_names'])
```

```python
y = pd.DataFrame(data=califorina['target'], columns=['Price'])
```

```python
df = pd.concat([X,y], axis = 1)
```

```python
df1 = df.copy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

regression = LinearRegression()

regression = regression.fit(X_train, y_train)

y_train_pred = regression.predict(X_train)

y_test_pred = regression.predict(X_test)

n_train = X_train.shape[0]
n_test = X_test.shape[0]
p = X_train.shape[1]


adj_r2_train = 1 - ((1 - r2_score(y_train, y_train_pred)) * (n_train-1)) / (n_train - p - 1)

adj_r2_test = 1 - ((1 - r2_score(y_test, y_test_pred)) * (n_test-1)) / (n_test - p - 1)


print("Train Data Set  : \n")
print(f"Mean Squared Error of linear regression for train dataset: {mean_squared_error(y_train, y_train_pred)}")
print(f"Mean Absolute Error of linear regression for train dataset: {mean_absolute_error(y_train, y_train_pred)}")
print(f"Root Mean Squared Error of linear regression for train dataset: {np.sqrt(mean_absolute_error(y_train, y_train_pred))}")
print(f"r2 score of train dataset : {r2_score(y_train, y_train_pred)}")
print(f"Adjusted R Square train dataset : {adj_r2_train}")
print("-" *100)
print("Test Data Set  : \n")
print(f"Mean Squared Error of linear regression for test dataset: {mean_squared_error(y_test, y_test_pred)}")
print(f"Mean Absolute Error of linear regression for test dataset: {mean_absolute_error(y_test, y_test_pred)}")
print(f"Root Mean Squared Error of linear regression for test dataset: {np.sqrt(mean_absolute_error(y_test, y_test_pred))}")
print(f"r2 score of test dataset : {r2_score(y_test, y_test_pred)}")
print(f"Adjusted R Square test dataset : {adj_r2_test}")
```
